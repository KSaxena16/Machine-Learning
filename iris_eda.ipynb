```python
count, bin_edges = np.histogram(iris_setosa["pl"],bins=10, density=True)
pdf = count/(sum(count))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

count, bin_edges = np.histogram(iris_versicolor["pl"],bins=10, density=True)
pdf = count/(sum(count))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

count, bin_edges = np.histogram(iris_virginica["pl"],bins=10, density=True)
pdf = count/(sum(count))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

plt.show()
```

```python
#Mean, Variance, Std-deviation,  
print("Means:")
print(np.mean(iris_setosa["pl"]))
#Mean with an outlier.
print(np.mean(np.append(iris_setosa["pl"],50)));
print(np.mean(iris_virginica["pl"]))
print(np.mean(iris_versicolor["pl"]))

print("\nStd-dev:");
print(np.std(iris_setosa["pl"]))
print(np.std(iris_virginica["pl"]))
print(np.std(iris_versicolor["pl"]))
```

```python
#Median
print("\nMedians:")
print(np.median(iris_setosa["pl"]))
#Median with an outlier
print(np.median(np.append(iris_setosa["pl"],50)));
print(np.median(iris_virginica["pl"]))
print(np.median(iris_versicolor["pl"]))
```

```python
#Quantiles
print("\nQuantiles:")
print(np.percentile(iris_setosa["pl"],np.arange(0, 100, 25)))
print(np.percentile(iris_virginica["pl"],np.arange(0, 100, 25)))
print(np.percentile(iris_versicolor["pl"], np.arange(0, 100, 25)))
#Percentiles
print("\n90th Percentiles:")
print(np.percentile(iris_setosa["pl"],90))
print(np.percentile(iris_virginica["pl"],90))
print(np.percentile(iris_versicolor["pl"], 90))
```

```python
#Median Absolute Deviation(similar to Std Dev)
from statsmodels import robust
print ("\nMedian Absolute Deviation")
print(robust.mad(iris_setosa["pl"]))
print(robust.mad(iris_virginica["pl"]))
print(robust.mad(iris_versicolor["pl"]))
```
